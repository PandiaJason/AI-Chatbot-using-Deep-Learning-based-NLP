```
# %%writefile chatbot.py
def chatbot():
    """
    main code of the chatbot.
    """
    global model,tokenizer, app, context, tce_info, self_info
    try:
        app = Flask(__name__)
        run_with_ngrok(app)
 
        #starts ngrok when the app is run
        @app.route("/")
        def index():
            """
            open wi and dispay html.
            """
            return render_template("index.html")
 
        @app.route("/get")
        def get_bot_response():
            """
            bot response
            """
            #get the user question
            question = request.args.get('msg')
            bert_needed=False
            response = duckduckgo.query(question)
            print('Response Type:',response.type)
            #print(response.json)
            global context
 
            if 'context' in question and 'web' in question:
                context = 'web'
                return str('Setting Context web')
 
            if 'context' in question and 'tce' in question:
                context = 'tce'
                return str('Setting Context tce')   
            
 
            if 'you' in question or 'your' in question:
                encoding = tokenizer.encode_plus(question, self_info, return_tensors="pt")
                input_ids = encoding["input_ids"]
                attention_mask = encoding["attention_mask"]
                start_scores, end_scores = model(input_ids, attention_mask=attention_mask)
                all_tokens = tokenizer.convert_ids_to_tokens(input_ids[0].tolist())
                answer_tokens = all_tokens[torch.argmax(start_scores) :torch.argmax(end_scores)+1]
                answer0 = tokenizer.decode(tokenizer.convert_tokens_to_ids(answer_tokens))
                return str(answer0)
            
            if context == 'tce':
                encoding = tokenizer.encode_plus(question, tce_info, return_tensors="pt")
                input_ids = encoding["input_ids"]
                attention_mask = encoding["attention_mask"]
                start_scores, end_scores = model(input_ids, attention_mask=attention_mask)
                all_tokens = tokenizer.convert_ids_to_tokens(input_ids[0].tolist())
                answer_tokens = all_tokens[torch.argmax(start_scores) :torch.argmax(end_scores)+1]
                answer1 = tokenizer.decode(tokenizer.convert_tokens_to_ids(answer_tokens))
                return str(answer1)
            if context == 'web':
 
                if question == 'tell me a joke':
                    joke=pyjokes.get_joke(language='en', category= 'neutral')
                    return str(joke)
                if response.type=='article':
                    print('Not Yet Procecced')
                elif response.type=='disambiguation':
                    answer_text=str(response.related[0].text)
                    print(answer_text)
                elif response.type=='category':
                    print('Not Yet Procecced')
                elif response.type=='name':
                    print('Not Yet Procecced')
                elif response.type=='exclusive':
                    #print(response.json)
                    #answer_text=response.answer.text
                    print("Sorry! I am a Chatbot. Not a Calculator.")
                    return str("Sorry! I am a Chatbot. Not a Calculator.")
                elif response.type=='nothing':
                    #print(r.json)
                    bert_needed=True
                    url = str(duckduckgo.get_zci(question))
                    print(url)
                    #url = 'ddldzvdzvn'
                    link = url
                    if link == url:
                        try:
                            wiki = url.rsplit('/', 1)[-1]
                            print(wiki)
                            answer_text = wikipedia.summary(wiki)
                            print("Wikipedia")
                            print(answer_text)
                        except:
                            page = requests.get(url).text
                            print("DuckDuckgo")
                            # Turn page into BeautifulSoup object to access HTML tags
                            soup = BeautifulSoup(page)
                            # Get text from all <p> tags.
                            p_tags = soup.find_all('p')
                            # Get the text from each of the “p” tags and strip surrounding whitespace.
                            p_tags_text=[tag.get_text().strip() for tag in p_tags]
                            # Filter out sentences that contain newline characters
                            # '\n' or don't contain periods.
                            sentence_list = [sentence for sentence in p_tags_text if not '\n' in sentence]
                            sentence_list = [sentence for sentence in sentence_list if '.' in sentence]
                            # Combine list items into string.
                            article = ' '.join(sentence_list)
                            answer_text = summarize(article, word_count = 500)
                            # ratio=0.3
                            if (len(answer_text)==0):
                                bert_needed=False
                                answer_text="Sorry, But You can  Check out this Link: " +url
                                #print(answer_text)
                elif response.type=='answer':
                    answer_text=duckduckgo.get_zci(question)
                    print(answer_text)
                else:
                    print('Response Type: Unknown')
                    answer_text=str(duckduckgo.get_zci(question))
                if bert_needed==True:
                    #return (str('Thinking.Please Wait'))
                    encoding = tokenizer.encode_plus(question, answer_text, return_tensors="pt")
                    input_ids = encoding["input_ids"]
                    attention_mask = encoding["attention_mask"]
                    start_scores, end_scores = model(input_ids, attention_mask=attention_mask)
                    all_tokens = tokenizer.convert_ids_to_tokens(input_ids[0].tolist())
                    answer_tokens = all_tokens[torch.argmax(start_scores) :torch.argmax(end_scores)+1]
                    answer = tokenizer.decode(tokenizer.convert_tokens_to_ids(answer_tokens))
                    if answer=='':
                        return (str("Sorry, But You can  Check out this Link: "+ url))
                    return str(answer)
                else:
                    return str(answer_text)
 
    except NameError:
        print()
 
    try:
        if __name__ == "__main__":
            app.run()
    except NameError:
        print()
chatbot()
```
